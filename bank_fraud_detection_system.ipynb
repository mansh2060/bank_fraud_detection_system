```python
#getting correlation matrix
num_columns=[]
for columns in sample_df.columns:
  if sample_df[columns].dtype != object :
    num_columns.append(columns)
plt.figure(figsize=(15,10))
correlation_matrix=sample_df[num_columns].corr()
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f',
            linewidth=0.5)
plt.title('Correlation Matrix')
#plt.xticks(rotation=45,ha='right')
#plt.yticks(rotation=45,ha='right')
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
label_encoder=LabelEncoder()
columns=['payment_type','employment_status','housing_status','source','device_os']
for col in columns:
  sample_df[col]=label_encoder.fit_transform(sample_df[col])
sample_df.head()
```

```python
sample_df.isnull().sum()
```

```python
sample_df.info()
```

```python
x=sample_df.drop(['fraud_bool'],axis=1)
y=sample_df['fraud_bool']
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#feature scaling
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
x_test=scaler.transform(x_test)
```

```python
import numpy as np
np.isnan(X_train).any()
```

```python
rows_with_nan = np.where(np.isnan(X_train))[0]
print("Rows with NaN:", rows_with_nan)
```

```python
cols_with_nan = np.where(np.isnan(X_train))[1]
print("Columns with NaN:", cols_with_nan)
```

```python
# Count total NaN values in the array
print("Total NaN values:", np.isnan(X_train).sum())

# Count NaN values per column
nan_per_column = np.isnan(X_train).sum(axis=0)
print("NaN values per column:", nan_per_column)
```

```python
#X_train = pd.DataFrame(X_train).fillna(0).values
```

```python
print(np.isinf(X_train).any())  # Check for infinite values
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
logistic_regression=LogisticRegression()
logistic_regression.fit(X_train,y_train)
y_pred=logistic_regression.predict(x_test)
```

```python
#accuracy
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
print("Classification report    :",classification_report(y_test,y_pred))
```

```python
print("Confusion Matrix     :",confusion_matrix(y_test,y_pred))
```
